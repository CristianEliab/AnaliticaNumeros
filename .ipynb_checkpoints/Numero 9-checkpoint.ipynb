Se observaron cuantas veces se encuentra el número 7 en el dataset de imágenes de sklearn.datasets. 

```python
from sklearn.datasets import load_digits
import pylab as plt
import numpy as np
```

```python
digits = load_digits()
numImagenes = len(digits.images) # Numero de imagenes, len es un método que provee el tamaño del arreglo
y = digits.target # el método nos provee las etiquetas de las imágenes en un arreglo
numero = 9; # El número a buscar

# Método que devuelve el total de números en el dataset
def numeroTotal(num,totalImagenes):
    numY = 0;
    for i in range(totalImagenes):
        if(y[i] == num):
            numY += 1;
    return numY;

# Método que devuelve todas las imágenes del número pasado por parametro en el dataset
def visualizarTodosNumeros(num,totalImagenes):
    for i in range(totalImagenes):
        if(y[i] == num):
            plt.imshow(digits.images[i], cmap=plt.cm.gray_r)
            plt.title('Pred:' + str(i))
            plt.show()

def visualizacion_2():
    images_and_predictions = list(zip(digits.images, y))
    for index, (image, prediction) in enumerate(images_and_predictions[:4]):
        plt.matshow(digits.images[index]) # Imprimir una imagen dada
        print(digits.images[index])
        plt.title('Pred: %i' % prediction)
        plt.show()
```

```python
# Mostrar la información: 

print('Número total de imágenes: ' + str(numImagenes));
print('Número total de 7 en el dataset: ' + str(numeroTotal(numero,numImagenes)));
```

```python
# Método que devuelve todas las imágenes del número pasado por parametro en el dataset

visualizarTodosNumeros(numero,numImagenes)
```

```python
# Método que calcula el promedio del oscuridad de la imagen que le enviaron por parametro
def promedio(imagen):
    num_pixel = 0;
    total = 0;
    for i in range(8):
        for j in range(8):
            if(imagen[i][j]!=0):
                num_pixel += imagen[i][j];
                total += 1;
    promedio = num_pixel//total;
    return promedio;
```

```python
# Método que reconstruye la imagen con el promedio de oscuridad que esta posee
def modificarImagen(promedio,imagen):
    for i in range(0,8):
        for j in range(0,8):
            if(imagen[i][j]>=promedio):
                imagen[i][j]=16;
            else:
                imagen[i][j]=0;
    return imagen;

# Método que compara las casillas de la imagen con la pantilla del número
def compararMatrices(promedio, imagen):
    acerto = False;
    modificarImagen(promedio,imagen);
    if(compararCasillas(imagen)==True):
        acerto = True;
    return acerto;
    
# Método que engloba la detección de las imagenes enviadas por parametro
def detectar_numero(imagen):
    acerto = False;
    p = promedio(imagen);
    if(compararMatrices(p,imagen) == True):
        acerto = True;
    return acerto;
```

```python
# Método que reconstruye la imagen con el promedio de oscuridad que esta posee
def compararCasillas(imagen):
    acerto = False;
    aciertos = 0;
    for i in range(0,8):
        if(i>=3 and i<=5):
            if(imagen[0][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=2 and i<=6):
            if(imagen[1][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i==2):
            if(imagen[2][i] == 16):
                aciertos += 1;
        elif(i>=4 and i<=6):
            if(imagen[2][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=5 and i<=6):
            if(imagen[3][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=2 and i<=6):
            if(imagen[4][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=2 and i<=6):
            if(imagen[5][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=3 and i<=4):
            if(imagen[6][i] == 16):
                aciertos += 1;
    for i in range(0,8):
        if(i>=2 and i<=5):
            if(imagen[7][i] == 16):
                aciertos += 1;
    
    if(aciertos >= 15):
        acerto = True;
    
    return acerto;
```

```python
# Método principal el cual verifica la identidad de la imagen y la compara si esta es un 7
def modeloMecanico(numero, totalImagenes):
    numAciertos = 0;
    for i in range(totalImagenes):
        if((detectar_numero(digits.images[i]) == True) and (y[i] == numero)):
            numAciertos += 1;
    return numAciertos;
```

```python
numero = 9;
total = numeroTotal(numero,numImagenes);
print("Número total de " + str(total));
print("Numero de dos encontrados correctamente: ", str(modeloMecanico(numero,numImagenes)))
```

```python
# Método para poder ver la modificaciones relizadas a las imagenes
def visualizacion_2():
    digits = load_digits();
    images_and_predictions = list(zip(digits.images, y))
    for index, (image, prediction) in enumerate(images_and_predictions):
        if(y[index] == 9):
            digits = load_digits();
            plt.matshow(digits.images[index])
            plt.title('Real')
            ##############################
            p = promedio(image);
            imagen = modificarImagen(p ,digits.images[index]);
            plt.matshow(imagen)
            plt.title('Modificada')
            plt.show()
        

visualizacion_2();
```
